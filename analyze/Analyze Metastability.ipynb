```python
import numpy as np
import matplotlib as plt
import scipy

import sys
sys.path.append('../utils')
sys.path.append('../dFC')
import widefield_utils
import connectivity_measures
```

```python
names = ["NXAK22.1A","NXAK14.1A","NXAK7.1B","NXAK4.1B","NRXN78.1D","NRXN78.1A"]
names_knock = ["NXAK24.1C","NXAK20.1B","NXAK16.1B","NXAK10.1A","NXAK4.1A","NRXN71.2A"]

meta = np.zeros(len(names))
avg_eig = np.zeros(len(names))
meta_knock = np.zeros(len(names))
avg_eig_knock = np.zeros(len(names))

for i,name in enumerate(names):
    
    filename = r"/home/k21208334/calcium_analyses/data/eigenvalue_timeseries/"+ name + "_eig1_timeseries.npy"
    eig_series = np.load(filename)
    meta[i] = np.std(eig_series)
    avg_eig[i] = np.mean(eig_series)

for i,name in enumerate(names_knock):
    
    filename = r"/home/k21208334/calcium_analyses/data/eigenvalue_timeseries/"+ name + "_eig1_timeseries.npy"
    eig_series = np.load(filename)
    meta_knock[i] = np.std(eig_series)
    avg_eig_knock[i] = np.mean(eig_series)
```

```python
plt.pyplot.plot(np.zeros(len(meta)),meta,'bo')
plt.pyplot.plot(np.ones(len(meta_knock)),meta_knock,'bo',color='red')
plt.pyplot.xlim(-1,2)
plt.pyplot.xticks([])
plt.pyplot.title('Spectral Metastability')
scipy.stats.mannwhitneyu(meta,meta_knock)
```

```python
plt.pyplot.plot(np.zeros(len(meta)),avg_eig,'bo')
plt.pyplot.plot(np.ones(len(meta_knock)),avg_eig_knock,'bo',color='red')
plt.pyplot.xlim(-1,2)
plt.pyplot.xticks([])
plt.pyplot.title('Average Spectral Radius')
scipy.stats.mannwhitneyu(avg_eig,avg_eig_knock)
```
